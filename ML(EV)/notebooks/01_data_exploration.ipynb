```python
from kagglehub import dataset_download
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = dataset_download("valakhorasani/electric-vehicle-charging-patterns")
print("Dataset downloaded to:", path)
```

```python
files = os.listdir(path)
print("Files inside dataset:", files)

# Load CSV (adjust filename if different)
df = pd.read_csv(os.path.join(path, files[0]))
print("Data shape:", df.shape)
print(df.head())
```

```python
print(df.info())
print(df.describe())
print("Missing values:\n", df.isnull().sum())
```

```python
# Distribution of charging energy
plt.figure(figsize=(8,5))
sns.histplot(df['Energy Consumed (kWh)'], bins=30, kde=True)
plt.title("Distribution of Charging Energy (kWh)")
plt.xlabel("Energy (kWh)")
plt.ylabel("Count")
plt.show()
```

```python
# Charging duration vs energy
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Charging Duration (hours)'], y=df['Energy Consumed (kWh)'])
plt.title("Charging Duration vs Energy Delivered")
plt.xlabel("Duration (hours)")
plt.ylabel("Energy Consumed (kWh)")
plt.show()
```
